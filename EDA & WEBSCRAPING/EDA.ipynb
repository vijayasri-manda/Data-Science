```python
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d.axes3d as p3
import numpy as np
```

```python
x=np.linspace(-5,5,100)
y=np.linspace(-5,5,100)
X,Y=np.meshgrid(x,y)
Z=np.sin(np.sqrt(X**2+Y**2))
```

```python
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
# Use the calculated X, Y, and z (lowercase) arrays for the surface plot
ax.plot_surface(X,Y,Z,cmap='viridis')
ax.set_xlabel('X-axis') # Corrected set_label to set_xlabel for better axis labeling
ax.set_ylabel('Y-axis') # Corrected set_label to set_ylabel
ax.set_zlabel('Z-axis') # Corrected set_label to set_zlabel
plt.show()
```

```python
x=np.linspace(-5,5,100)
y=np.linspace(-5,5,100)
X,Y=np.meshgrid(x,y)
Z=(X**2+Y**2)
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
ax.plot_surface(X,Y,Z,cmap='viridis')
ax.set_xlabel('X-axis') # Corrected set_label to set_xlabel for better axis labeling
ax.set_ylabel('Y-axis') # Corrected set_label to set_ylabel
ax.set_zlabel('Z-axis') # Corrected set_label to set_zlabel
plt.show()
```

```python
#Donut chart
plt.figure(figsize=(10,10))
labels=['python','java','c++','c','Ruby']
sizes=[10,20,30,40,50]
# Create a new figure and axes for the donut chart
fig, ax = plt.subplots()
ax.pie(sizes,labels=labels,startangle=90,wedgeprops={'width':0.4})
ax.axis('equal')
plt.title("Donut Chart")
plt.show()
```
